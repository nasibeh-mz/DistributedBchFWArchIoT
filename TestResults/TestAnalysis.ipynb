```python
def calculate_dif(list1, list2):
    difs = []
    difs_total = 0
    for elem in range(0,len(list2)):
        difs.append(list2[elem] - list1[elem])
        difs_total += difs[elem]
    average = difs_total / len(list2)
    return difs, average
```
